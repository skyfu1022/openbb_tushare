# openbb_tushare

Check openbb_tushare version

```python
import os
from dotenv import load_dotenv
from openbb import obb

if 'info' in obb.reference:
    obj = obb.reference["info"]["extensions"]["openbb_provider_extension"]
    print([item for item in obj if "tushare" in item])

symbols = "601919,600519,000001,00300"
symbols_a = "601919,600519,000001,600325,600028,600036,430047"
symbol_a = "600325.SH"
symbol_hk = "01658"
provider = "tushare"
start_date="2025-05-01"
end_date="2025-05-27"
start_date_ts="20250501"
end_date_ts="20250527"

os.environ["TUSHARE_API_KEY"] = obb.user.credentials.tushare_api_key.get_secret_value()
load_dotenv() 
tushare_api_key = os.environ.get("TUSHARE_API_KEY")
if tushare_api_key is None:
    raise ValueError("TUSHARE_API_KEY environment variable not set.")
```

## 1. 股票相关模型 (Equity Models)
### 1.1 Equity Historical Price (股票历史价格)

```python
obb.equity.price.historical(symbol=symbol_a, provider=provider, use_cache=True, start_date=start_date, end_date=end_date).to_dataframe().tail()
```

```python
import tushare as ts
pro = ts.pro_api(tushare_api_key)
df_data = pro.daily(ts_code=symbol_a, start_date=start_date_ts, end_date=end_date_ts)
df_data = df_data.sort_values('trade_date')
df_data.tail(11)
```

#### 前复权行情

```python
qfq_df = ts.pro_bar(ts_code=symbol_a, adj='qfq', start_date=start_date_ts, end_date=end_date_ts)
qfq_df = qfq_df.sort_values('trade_date')
qfq_df.tail(11)
```

#### 后复权行情

```python
hfq_df = ts.pro_bar(ts_code=symbol_a, adj='hfq', start_date=start_date_ts, end_date=end_date_ts)
hfq_df = hfq_df.sort_values('trade_date')
hfq_df.tail(11)
```
